# Performing Feature Engineering

```python
# Perform one-hot encoding
df = pd.get_dummies(df, columns = ['Status'])
```

```python
df.isnull().sum()
```

```python

# Fill missing values with the mean of their respective columns
df = df.fillna(df.mean(numeric_only=True))

# Verify that missing values have been filled
print("Total null :",df.isnull().sum().sum())
```

```python
# Create train and test data
X = df.drop(columns = ['Life expectancy '])
y = df[['Life expectancy ']]
```

```python
# Convert the data type to float32
X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
# spliting the data into training, testing and validation sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor

# Initialize the Gradient Boosting regressor
model = GradientBoostingRegressor(learning_rate=0.1, max_depth=3, n_estimators=100, random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Compute metrics
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Adjusted R-squared
k = X_test.shape[1]
n = len(X_test)
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)

# Print results
print('RMSE = {:.3f}'.format(RMSE))  
print('MSE = {:.3f}'.format(MSE))
print('MAE = {:.3f}'.format(MAE))
print('R2 = {:.3f}'.format(r2))
print('Adjusted R2 = {:.3f}'.format(adj_r2))
```

## Explanation of Evaluation Metrics

### RMSE (Root Mean Squared Error)
- **RMSE = 2.173**
  - RMSE measures the average magnitude of the errors in predicting life expectancy. A lower RMSE indicates better accuracy. Here, the RMSE of 2.173 means, on average, the model's predictions deviate by approximately 2.173 years from the actual values.

### MSE (Mean Squared Error)
- **MSE = 4.721**
  - MSE is the average of the squared differences between predicted and actual values. It provides a measure of the model's prediction error. A lower MSE indicates better accuracy. The MSE of 4.721 highlights the average squared deviation of predictions from actual values.

### MAE (Mean Absolute Error)
- **MAE = 1.561**
  - MAE measures the average absolute differences between predicted and actual values. It is less sensitive to outliers compared to RMSE. Here, an MAE of 1.561 means the model's predictions are off by approximately 1.561 years on average.

### R-squared (R2)
- **R2 = 0.949**
  - R-squared is a measure of how well the model explains the variance in the dependent variable (life expectancy). A higher R2 indicates a better fit of the model to the data. An R2 of 0.949 suggests that 94.9% of the variance in life expectancy is explained by the features in the model.

### Adjusted R-squared
- **Adjusted R2 = 0.948**
  - Adjusted R-squared adjusts R2 for the number of predictors in the model. It penalizes model complexity. A high adjusted R2 (0.948) indicates that the selected features adequately explain the variance in life expectancy without overfitting.

## Conclusion
The regression model demonstrates strong performance in predicting life expectancy based on the provided dataset. The low values of RMSE, MSE, and MAE indicate that the model's predictions are close to the actual values. Additionally, the high R-squared (0.949) and adjusted R-squared (0.948) scores confirm the model's robustness in explaining the variability in life expectancy. Therefore, the model can be considered reliable for predicting life expectancy using the chosen features.
